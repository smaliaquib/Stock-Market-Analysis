<h3>1.4 In a separate array , calculate the monthwise VWAP</h3>

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
by = df.groupby(['Year','Month'])
```

```python
# 2017 VWAP
li_2017=[]
for i in range(5,13):
    li_2017.append(sum(by.get_group((2017,i))['Average Price']*by.get_group((2017,i))['Total Traded Quantity'])/sum(by.get_group((2017,i))['Total Traded Quantity']))
```

```python
# 2018 VWAP
li_2018=[]
for i in range(1,13):
    li_2018.append(sum(by.get_group((2018,i))['Average Price']*by.get_group((2018,i))['Total Traded Quantity'])/sum(by.get_group((2018,i))['Total Traded Quantity']))
```

```python
# 2019 VWAP
li_2019=[]
for i in range(1,6):
    li_2019.append(sum(by.get_group((2019,i))['Average Price']*by.get_group((2019,i))['Total Traded Quantity'])/sum(by.get_group((2019,i))['Total Traded Quantity']))
```

<h3>1.5 average price over the last N days of the stock price data where N is a user defined parameter</h3>

```python
def average(inp):
    avg=0
    for i in range(inp):
        avg = avg + df['Average Price'][i]
    print("Average Price = {0}".format(avg/inp))
```

```python
def pl(inp):
    pro=0
    p=0
    los=0
    l=0
    for i in range(inp):
        cp = df['Prev Close'][i]
        sp = df['Close Price'][i]

        if(cp > sp):
            loss = cp - sp
            loss_percent = (loss / cp) * 100
            l = l+1
            los = los + loss_percent
        elif(sp > cp):
            profit = sp - cp
            profit_percent = (profit / cp) * 100
            p = p+1
            pro = pro + profit_percent
            
    print("Profit percent = {0} %".format(pro/p*10))
    print("Loss percent = {0} %".format(los/l*10))
```

```python
inp = int(input())
pl(inp)
average(inp)
```

<h3>1.6 Add a column 'Day_Perc_Change' where the values are the daily change in percentages</h3>

```python
df['Day_Perc_Change']=df['Close Price'].pct_change() *100
```

```python
df.drop(0,axis=0,inplace=True)
```

```python
df.loc[(df['Day_Perc_Change'] > -0.5) & (df['Day_Perc_Change'] < 0.5),'Trend'] = 'Slight or No change'
df.loc[(df['Day_Perc_Change'] > 0.5) & (df['Day_Perc_Change'] < 1),'Trend'] = 'Slight positive'
df.loc[(df['Day_Perc_Change'] > -1) & (df['Day_Perc_Change'] < -0.5),'Trend'] = 'Slight negative'
df.loc[(df['Day_Perc_Change'] > 1) & (df['Day_Perc_Change'] < 3),'Trend'] = 'Positive'
df.loc[(df['Day_Perc_Change'] > -3) & (df['Day_Perc_Change'] < -1),'Trend'] = 'Negative'
df.loc[(df['Day_Perc_Change'] > 3) & (df['Day_Perc_Change'] < 7),'Trend'] = 'Among top gainers'
df.loc[(df['Day_Perc_Change'] > -7) & (df['Day_Perc_Change'] < -3),'Trend'] = 'Among top losers'
df.loc[(df['Day_Perc_Change'] > 7),'Trend'] = 'Bull run'
df.loc[(df['Day_Perc_Change'] < -7),'Trend'] = 'Bear drop'
```
